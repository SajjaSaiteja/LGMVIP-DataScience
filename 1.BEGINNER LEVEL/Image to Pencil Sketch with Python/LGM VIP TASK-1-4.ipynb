```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading the Image
```

```python
img = cv2.imread("fruits.jpg")
cv2.imshow("Apples", img)
cv2.waitKey(0)
```

```python
#Original Image to Grayscale Image Conversion
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap="gray")
plt.title("Apples Image in Black And White")
```

```python
#Grayscale Image to Inverted Image
```

```python
inverted = 255 - gray
plt.imshow(inverted,cmap="gray")
plt.title("Apples Image in Black And White")
```

```python
#GaussianBlur of the Inverted Image
```

```python
blur = cv2.GaussianBlur(inverted, (21, 21), 0)
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred,cmap="gray")
plt.title("Inverted Blurred GrayScale Apples Image")
```

```python
#Pencil Sketch of the Gaussian Blur Image
```

```python
pencil_sketch = cv2.divide(gray, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Inverted Blurred GrayScale Apples Image")
```
